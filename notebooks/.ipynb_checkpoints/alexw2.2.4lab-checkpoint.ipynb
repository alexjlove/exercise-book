```python
import numpy
import pandas
billboard = pandas.read_csv('CSV/billboard.csv')
```

```python
billboard
```

```python
new_column_names = []
for column in billboard.columns:
    if column.endswith('week') and len(column)==9:
        good_name = 'x0' + column[1:]
        new_column_names.append(good_name)
    else:
        new_column_names.append(column)
new_column_names
```

```python
billboard.columns = new_column_names
```

```python
def time_conv(x):
    time2 = []
    for i in range(len(x)):
        if ':' in x[i]:
            sp = x[i].split(':')
            new = float(sp[0])*60 + float(sp[1])
            time2.append(new)
        else:
             time2.append(float(x[i])*60)
    return time2
```

```python
billboard['time2'] = time_conv(billboard.time)
```

```python
billboard[billboard.year == 2000].time2.mean()
```

```python
billboard.time2.mode()
```

```python
numpy.std(billboard.time2)
```

```python
print billboard.time2.max()
print billboard.time2.min()
print (billboard.time2.max() - billboard.time2.min())
```

```python
billboard.groupby('genre').time2.mean()
```
