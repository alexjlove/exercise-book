```python
sample_text = """
This is a sentence. These sentences are repetitive. But they use different words.
This is another sentence. It uses the same words. Isn't this repetitive?
"""
```

```python
nltk.sent_tokenize(sample_text)
```

```python
word_counts = []
words_used = set()
for sent in nltk.sent_tokenize(sample_text):
    words = [x for x in nltk.word_tokenize(sent) if x not in ['.', '?']]
    word_counts.append(len(words))
    for w in words:
        words_used.add(w.upper())
    print words
print "Word counts per sentence",word_counts
print "Word count overall", sum(word_counts)
print "Distinct words", len(words_used)
print "Lexical diversity =", 1.0 * len(words_used) / sum(word_counts)
```

# Stemming

```python
import nltk.stem.porter
```

```python
import nltk.stem.snowball
```

```python
snowball = nltk.stem.snowball.SnowballStemmer("english")
```

```python
snowball.stem("running")
```

```python
porter = nltk.stem.porter.PorterStemmer()
porter.stem("running")
```

```python
word_counts = []
words_used = set()
word_stems = set()
for sent in nltk.sent_tokenize(nltk.corpus.gutenberg.raw('carroll-alice.txt')):
    words = [x for x in nltk.word_tokenize(sent) if x not in ['.', '?']]
    word_counts.append(len(words))
    for w in words:
        word_stems.add(snowball.stem(w.lower()))
        words_used.add(w.upper())
#print "Word counts per sentence",word_counts
print "Word count overall", sum(word_counts)
print "Distinct words", len(words_used)
print "Distinct word stems", len(word_stems)
print "Lexical diversity =", 1.0 * len(words_used) / sum(word_counts)
print "Lexical stem diversity =", 1.0 * len(word_stems) / sum(word_counts)
```

```python
for s in nltk.sent_tokenize(nltk.corpus.gutenberg.raw('carroll-alice.txt')):
    was = False
    words = nltk.word_tokenize(s)
    tagged = nltk.pos_tag(words)
    for (word, part_of_speech) in tagged:
        if part_of_speech == 'VBD' and word == 'was':
            was = True
            continue
        if was and part_of_speech == 'JJ':
            print "was",word
            continue
        if part_of_speech == 'VBD':
            print word
```

```python
import pandas
alice = nltk.word_tokenize(nltk.corpus.gutenberg.raw('carroll-alice.txt'))
bible = nltk.word_tokenize(nltk.corpus.gutenberg.raw('bible-kjv.txt'))
word_counts = pandas.Series(
  data = [len(alice), len(bible)],
  index = ['Alice', "Bible"]
)
nlp_df = pandas.DataFrame(
  { 'word_counts': word_counts}
)
nlp_df
```

```python
references_to_alice = pandas.Series(
  data = [len([x for x in alice if x == 'Alice']),
          len([x for x in bible if x == 'Alice'])
          ],
  index = ['Alice', "Bible"]
)
nlp_df['Alice_count'] = references_to_alice
nlp_df
```

```python
references_to_god = pandas.Series(
  data = [len([x for x in alice if x == 'God']),
          len([x for x in bible if x == 'God'])
          ],
  index = ['Alice', "Bible"]
)
nlp_df['God_count'] = references_to_god
nlp_df
```

```python
references_to_the = pandas.Series(
  data = [len([x for x in alice if x == 'the']),
          len([x for x in bible if x == 'the'])
          ],
  index = ['Alice', "Bible"]
)
nlp_df['The_count'] = references_to_the
nlp_df
```

# Simple sentiment analysis

```python
sentences = [
 "I like chocolate",
 "I like frogs",
 "I hate cloudy days",
 "I hate cockroaches",
 "The weather today is fine",
 "I am enjoying this class"
]

sentiment = [
  0.6,
  0.6,
  -0.6,
  -0.8,
  0.5,
  0.9
]

about_animals = [
  False,
  True,
  False,
  True,
  False,
  False
]

X = tfidf.fit_transform(sentences)
X
```

```python
tfidf.get_feature_names()
```

```python
import sklearn.linear_model
ols = sklearn.linear_model.LinearRegression()
ols.fit(X, sentiment)
zip(tfidf.get_feature_names(), ols.coef_)
```

```python
ols.predict(tfidf.transform(["I like cockroaches"]))
```

```python
logreg = sklearn.linear_model.LogisticRegression()
logreg.fit(X, about_animals)
zip(tfidf.get_feature_names(), logreg.coef_[0])
```

```python
logreg.predict(tfidf.transform(["I am enjoying frogs"]))
```

```python
logreg.predict_proba(tfidf.transform(["I am enjoying frogs"]))
```

```python
logreg.predict(tfidf.transform(["I am enjoying cloudy days"]))
```

```python
logreg.predict_proba(tfidf.transform(["I am enjoying cloudy days"]))
```
