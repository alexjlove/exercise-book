What we are wanting to do is collect everything up until the
first [ or (

So we need a regular expression that matches anything except for these
characters.

A regexp for matching just a letter is [A-Z]. A regexp for matching
a number is [0-9].

A regexp for matching anything other than a number is [^0-9]

A regexp for matching [ or ( only would be [[(], but we want the opposite,
so we want the regexp:

   [^[(]
   
And we want as many of these characters in a row that we can find, so
we add a * at the end.

   [^[(]*
   
Pandas captures the first group marked with (parentheses).

Thus...

   ([^[(]*)
  

```
ugly_country_names.str.extract('([^[(]*)')
```

```
populations = dfs[2].copy()
populations['Name'] = ugly_country_names.str.extract('([^[(]*)').str.strip()
```

# Option 1

Repair after the fact.

We will use an outer join instead of an inner join.

```
animals = pandas.DataFrame(
  data = {'name': ['Chong', 'Alex', 'Byron', 'James'],
          'fav_animal': ['frog', 'panda', 'dog', 'dog']
          }
)
```

```
animals
```

```
favourite_colours = pandas.DataFrame(
  data = {'name': ['Guarav', 'James', 'Chong', 'Alex'],
          'fav_colour': ['Green', 'Blue', 'Orange', 'Blue']}
)
favourite_colours
```

With a **left inner** join (the default), we only get fields that exist in both dataframes.

```
animals.merge(favourite_colours, on='name')
```

With a **left outer** join, we get all of the left hand dataframe,
appended with whatever from the right-hand.

```
animals.merge(favourite_colours, on='name', how='left')
```

```
drinks_outer = drinks.merge(populations, 
                            left_on='country',
                            right_on='Name',
                            how='left'
                            ).copy()
```

```
len(drinks_outer)
```

```
len(drinks)
```

```
drinks_outer[drinks_outer.Population.isnull()].country
```

... and so on for the other 16 countries, figure out their indexes by hand...

```
mappings = {
'USA': "United States",
'St. Lucia': "Saint Lucia",
'Micronesia': "Federated States of Micronesia"
}
```

```
for (drinks_table_country_name, wikipedia_country_name) in mappings.iteritems():
    idx = drinks_outer[drinks_outer.country == drinks_table_country_name].index[0]
    pop = populations.Population[populations.Name == wikipedia_country_name].median()
    drinks_outer.loc[idx,'Population'] = pop
```

```
drinks_outer[drinks_outer.Population.isnull()].country
```

# Option 2 - Fix the drinks table

```
def correct_wrong_drinks_table_name(x):
    if x in mappings:
        return mappings[x]
    return x
```

```
correct_wrong_drinks_table_name('USA')
```

```
correct_wrong_drinks_table_name('Australia')
```

```
drinks['Option2_Corrected_Name'] = drinks.country.map(correct_wrong_drinks_table_name)
```

```
drinks
```

```
option2_drinks = drinks.merge(populations,
                              left_on='Option2_Corrected_Name',
                              right_on='Name')
```
