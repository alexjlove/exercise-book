### Section I: Import the data

```python
%matplotlib inline 

import pandas as pd
import seaborn as sns
from sklearn import metrics
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
mtcars = pd.read_csv("CSV/mtcars.csv")
mtcars.head()
```

### Plot the Data

```python
sns.pairplot(mtcars, hue='cyl')
```

### Extract the features you want to use in clustering into a matrix

```python
df = mtcars.drop('Car', axis=1)
```

### Cluster the data using K-Means Clustering

Cluster two of the variables of your choice. Choose K based on your plots and the behavior of the data

```python
X = df[['disp', 'wt']]
k=3
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)
```

### Find the Silhoutte Score and plot the features and clusters

```python
clusters = kmeans.cluster_centers_
metrics.silhouette_score(X, kmeans.labels_, metric='euclidean')
X['labels'] = kmeans.labels_
```

```python
sns.lmplot(x='disp', y='wt', hue='labels', data = X, fit_reg=False)
```
