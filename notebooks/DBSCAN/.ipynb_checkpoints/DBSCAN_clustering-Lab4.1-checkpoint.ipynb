```python
import pandas as pd
import numpy as np

from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline 
```

## 1. Load the data

```python
airport = pd.read_csv('../CSV/airport2.csv')
airport.head()
```

```python
X = airport.ix[:,2:].as_matrix()
y = airport.ix[:,0]
```

## 2. Format and graph the data

#### 2.1 Plot the data

```python
plt.scatter(X[:,2], X[:,3])
```

```python
plt.scatter(X[:,0], X[:,3])
```

#### 2.2 Select columns of interest and standardize the data

```python
X_stan = StandardScaler().fit_transform(X)
```

## 3. DBSCAN

#### 3.1 Setup the DBSCAN from scikit

```python
dbscan = DBSCAN(eps=0.5, min_samples=3).fit(X_stan)
```

#### 3.2 Calculate the Labels

```python
labels = dbscan.labels_
```

#### 3.3 Calculate the Core Samples

```python
core_samples = np.zeros_like(labels, dtype = bool)
core_samples[dbscan.core_sample_indices_] = True
print(core_samples)
```

#### 3.4 Plot the resulting clusters

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
plt.scatter(X[:,0], X[:,3], c=dbscan.labels_, s=50)
```

Now, repeat the above steps for various levels of minimum points and epsilon and see how that affects the results of the DBSCAN algorithm 

```python
eps = [0.2, 0.8, 1.0]
min_p = [1, 3, 5]
for e in eps:
    for s in min_p:
        dbscan = DBSCAN(eps=e, min_samples=s).fit(X_stan)
        plt.scatter(X[:,0], X[:,3], c=dbscan.labels_, s=50)
        plt.show()
```

#### 3.5 Perform a K-means Clustering for comparison

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_stan)
```

```python
labels = kmeans.labels_
```
