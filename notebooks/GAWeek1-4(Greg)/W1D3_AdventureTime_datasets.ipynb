```python
import pandas
import matplotlib.pyplot
```

```python
drinking_habits = pandas.read_csv('national-drinking-habits.csv', index_col=0)
drinking_habits
```

```python
drinking_habits.head().plot.barh()
```

```python
beermean = drinking_habits.beer_servings.mean()
winemean = drinking_habits.wine_servings.mean()
spiritmean = drinking_habits.spirit_servings.mean()
beermean
```

```python
matplotlib.pyplot.scatter(drinking_habits.wine_servings, drinking_habits.total_litres_of_pure_alcohol)
```

```python
matplotlib.pyplot.scatter(drinking_habits.spirit_servings, drinking_habits.total_litres_of_pure_alcohol)
```

```python
sum_servings = drinking_habits.beer_servings + drinking_habits.wine_servings + drinking_habits.spirit_servings
beer_percent = drinking_habits.beer_servings / sum_servings
wine_percent = drinking_habits.wine_servings / sum_servings
spirit_percent = drinking_habits.spirit_servings / sum_servings
```

# Spirit and Beer consumption are inversely related

```python
matplotlib.pyplot.scatter(beer_percent, spirit_percent)
```

# Mean Alcohol Consumption

```python
labels = ['Beer Mean', 'Wine Mean', 'Spirit Mean']
sizes = [beermean, winemean, spiritmean]
colors = ['gold', 'red', 'blue']
explode = (0.1, 0, 0)

matplotlib.pyplot.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
```

# Bar Plot of Total Litres

```python
drinking_habits.total_litres_of_pure_alcohol.max()
```

```python
drinkers_selector = drinking_habits.total_litres_of_pure_alcohol != 0
drinkers = drinking_habits[drinkers_selector].copy()
A = drinkers.total_litres_of_pure_alcohol[drinkers.total_litres_of_pure_alcohol < 1.8].count()
B = drinkers.total_litres_of_pure_alcohol[(drinkers.total_litres_of_pure_alcohol < 3.6) & (drinkers.total_litres_of_pure_alcohol > 1.8)].count()
C = drinkers.total_litres_of_pure_alcohol[(drinkers.total_litres_of_pure_alcohol < 5.4) & (drinkers.total_litres_of_pure_alcohol > 3.6)].count()
D = drinkers.total_litres_of_pure_alcohol[(drinkers.total_litres_of_pure_alcohol < 7.2) & (drinkers.total_litres_of_pure_alcohol > 5.4)].count()
E = drinkers.total_litres_of_pure_alcohol[(drinkers.total_litres_of_pure_alcohol < 9.0) & (drinkers.total_litres_of_pure_alcohol > 7.2)].count()
F = drinkers.total_litres_of_pure_alcohol[(drinkers.total_litres_of_pure_alcohol < 10.8) & (drinkers.total_litres_of_pure_alcohol > 9.0)].count()
G = drinkers.total_litres_of_pure_alcohol[(drinkers.total_litres_of_pure_alcohol < 12.6) & (drinkers.total_litres_of_pure_alcohol > 10.8)].count()
H = drinkers.total_litres_of_pure_alcohol[(drinkers.total_litres_of_pure_alcohol < 14.5) & (drinkers.total_litres_of_pure_alcohol > 12.6)].count()
```

```python
s = pandas.Series(
    data=[A,B,C,D,E,F,G,H],
    index=["Teetolaller", "Sub-Teetolallers", "3", "4", "5", "6", "Sub-Alcoholics", "Alcoholics"]
    )
s.plot.bar()
```

# Bar Plot of Biggest Drinkers by Continent

```python
continent_gb = drinking_habits.groupby('continent').total_litres_of_pure_alcohol.sum()
```

```python
continent_gb.plot.bar()
```

```python
EU_selector = drinking_habits.continent == "EU"
EU = drinking_habits[EU_selector].copy()
AF_selector = drinking_habits.continent == "AF"
AF = drinking_habits[AF_selector].copy()
AS_selector = drinking_habits.continent == "AS"
AS = drinking_habits[AS_selector].copy()
OC_selector = drinking_habits.continent == "OC"
OC = drinking_habits[OC_selector].copy()
SA_selector = drinking_habits.continent == "SA"
SA = drinking_habits[SA_selector].copy()
matplotlib.pyplot.boxplot([EU.total_litres_of_pure_alcohol, AF.total_litres_of_pure_alcohol,
                          AS.total_litres_of_pure_alcohol, OC.total_litres_of_pure_alcohol,
                          SA.total_litres_of_pure_alcohol],
                          vert=True,
                          patch_artist=True,
                          )
xticks = matplotlib.pyplot.xticks([1, 2, 3, 4, 5], ['Europe', 'Africa', 'Asia', 'Oceania', 'SouthAmerica'])
```

# Popular Drinks by Continent

```python
beer_by_cont = drinking_habits.groupby('continent').beer_servings.sum()
wine_by_cont = drinking_habits.groupby('continent').wine_servings.sum()
spirit_by_cont = drinking_habits.groupby('continent').spirit_servings.sum()
```

```python
q = pandas.DataFrame(
    data = {"Beer": beer_by_cont, "Wine":wine_by_cont, "Spirit":spirit_by_cont}
    )
q.plot.bar(color=['gold','blue','red'])
```
