```python
import pandas as pd
import numpy as np
```

```python
# Read in realEstate.csv as a data frame
realEstate = pd.read_csv('CSV/Sacramentorealestatetransactions.csv')
realEstate
```

```python
# The simplest pivot table must have a dataframe and an 
# index . Let’s use the city as our index.
pd.pivot_table(realEstate, index=['city'])
```

```python
# How about indexing on multiple values. Let's look at the 
# data by city and zip next. 
pd.pivot_table(realEstate, index=['city', 'zip'])
```

```python
# What we really want to take a look at next is price. So, 
# let's explicitly define the 'price' column by 
# using the values field.
pd.pivot_table(realEstate, index=['city', 'zip'], values=['price', 'sq__ft'])
```

```python
# The 'price' column automatically averages the data.  
# We can do a count or a sum by using aggfunc and np.sum.
pd.pivot_table(realEstate, index=['city', 'zip'], values=['price'], aggfunc=sum)
```

```python
indexed_df = pd.pivot_table(realEstate, index=['city', 'zip'])
seaborn.boxplot(x='city', y='price', data=realEstate)
```
