# Sydney Weather

```python
weather = requests.get('http://www.bom.gov.au/nsw/forecasts/sydney.shtml')
weasoup = BeautifulSoup(weather.text, 'lxml')
```

```python
earliest_day_div = weasoup.find('div', class_='day main')
main_forecast = earliest_day_div.find('div', class_='forecast')
max_temp = earliest_day_div.find('em')
```

```python
max_temp.string
```

```python
next_days = weasoup.find_all('div', class_='day')
```

```python
next_days[0]
```

```python
next_days_max = []
next_days_min = []
for x in next_days:
    forecasts = x.find('div', class_='forecast')
    min_temp = forecasts.find('em', class_='min')
    max_temp = forecasts.find('em', class_='max')
    next_days_max.append(max_temp.string)
    #next_days_min.append(min_temp.string)
temp_forecasts = {"Today": int(next_days_max[0]), "Tomorrow":int(next_days_max[1]),
                  "Day 3": int(next_days_max[2]), "Day 4": int(next_days_max[3]),
                  "Day 5": int(next_days_max[4]), "Day 6": int(next_days_max[5]),
                  "Day 7": int(next_days_max[6])}
temp_forecasts
```
