## 1: Inspect the database

If we were connected via console, it would be easy to list all tables using `\dt`. We can however access table information performing a query on the `information_schema.tables` table.

### 1.a: List Tables

1. write a `SELECT` statement that lists all the tables in the public schema of the `northwind` database, sorted alphabetically

```python
%%sql

SELECT * FROM information_schema.tables
WHERE table_schema = 'public'
ORDER BY table_name ASC;
```

### 1.b: Print Schemas

The table `INFORMATION_SCHEMA.COLUMNS` contains schema information on each.

Query it to display schemas of all the public tables. In particular we are interested in the column names and data types. Make sure you only include public schemas to avoid cluttering your results with a bunch of postgres related stuff.

```python
%%sql

SELECT table_name, column_name, data_type FROM information_schema.columns
WHERE table_schema = 'public';
```

### 2.a: What categories of products is the company selling?

Remember that PostgreSQL is case sensitive.

```python
%%sql
SELECT "CategoryName" FROM categories;
```

### 2.b: How many products per category does the catalog contain?

Keep in mind that you can cast a %sql result to a pandas dataframe using the `.DataFrame()` method.

```python
%%sql
SELECT "CategoryName", COUNT("ProductID") FROM products
JOIN categories ON products."CategoryID" = categories."CategoryID"
GROUP BY categories."CategoryID";
```

### 2.c: How many not discontinued products per category?

```python
result = %sql SELECT "CategoryName", COUNT("ProductID") FROM products \
                JOIN categories ON products."CategoryID" = categories."CategoryID" \
                WHERE "Discontinued" = 0 \
                GROUP BY categories."CategoryID";
result
```

## 3: Orders

Now that we have a better understanding of products, let's start digging into orders.

- How many orders in total?
- How many orders per year
- How many orders per quarter
- Which country is receiving the most orders
- Which country is receiving the least
- What's the average shipping time (ShippedDate - OrderDate)
- What customer is submitting the highest number of orders?
- What customer is generating the highest revenue (need to pd.merge with order_details)

```python
%%sql
SELECT * FROM orders;
```

### 3.a: How many orders in total?

```python
%%sql
SELECT COUNT(*) FROM orders
```

### 3.b: How many orders per year?

```python
%%sql
SELECT COUNT(*),
    CASE
        WHEN "OrderDate" BETWEEN '1996-01-01' and '1996-12-31' THEN '1996'
        WHEN "OrderDate" BETWEEN '1997-01-01' and '1997-12-31' THEN '1997'
        WHEN "OrderDate" BETWEEN '1998-01-01' and '1998-12-31' THEN '1998'
    END "Year"
FROM orders
GROUP BY "Year";
```

### 3.c: How many orders per quarter?

Make a line plot for these.

```python
%%sql
SELECT COUNT(*),
    CASE
        WHEN "OrderDate" BETWEEN '1996-04-01' and '1996-06-30' THEN '1996 Q2'
        WHEN "OrderDate" BETWEEN '1996-07-01' and '1996-09-30' THEN '1996 Q3'
        WHEN "OrderDate" BETWEEN '1996-10-01' and '1996-12-31' THEN '1996 Q4'
        WHEN "OrderDate" BETWEEN '1997-01-01' and '1997-03-31' THEN '1997 Q1'
        WHEN "OrderDate" BETWEEN '1997-04-01' and '1997-06-30' THEN '1997 Q2'
        WHEN "OrderDate" BETWEEN '1997-07-01' and '1997-09-30' THEN '1997 Q3'
        WHEN "OrderDate" BETWEEN '1997-10-01' and '1997-12-31' THEN '1997 Q4'
        WHEN "OrderDate" BETWEEN '1998-01-01' and '1998-03-31' THEN '1998 Q1'
        WHEN "OrderDate" BETWEEN '1998-04-01' and '1998-06-30' THEN '1998 Q2'
        WHEN "OrderDate" BETWEEN '1998-07-01' and '1998-09-30' THEN '1998 Q3'
    END "Quarters"
FROM orders
GROUP BY "Quarters"
ORDER BY "Quarters" ASC;
```

### 3.d: Which country is receiving the most orders?

```python
%%sql
SELECT "ShipCountry", COUNT(*) FROM orders
GROUP BY "ShipCountry"
ORDER BY "count" DESC
LIMIT 5;
```

### 3.e: Which country is receiving the least?

```python
%%sql
SELECT "ShipCountry", COUNT(*) FROM orders
GROUP BY "ShipCountry"
ORDER BY "count" ASC
LIMIT 5;
```

### 3.f: What's the average shipping time (ShippedDate - OrderDate)?

```python
result = %sql SELECT "ShippedDate", "OrderDate" FROM orders;
df = result.DataFrame()
df['ShippedDate'] = pd.to_datetime(df.ShippedDate)
df['OrderDate'] = pd.to_datetime(df.OrderDate)
df['ShipTime'] = df.ShippedDate - df.OrderDate
df.ShipTime.mean()
```

### 3.g: What customer is submitting the highest number of orders?

```python
%%sql
SELECT COUNT(*), "CompanyName" FROM orders
JOIN customers ON customers."CustomerID" = orders."CustomerID"
GROUP BY customers."CompanyName"
ORDER BY count DESC
LIMIT 5;
```

### 3.h: What customer is generating the highest revenue (need to pd.merge with order_details)?

```python
result = %sql SELECT "OrderID", "UnitPrice", "Quantity", "Discount" FROM order_details;
orderdf = result.DataFrame()

result2 = %sql SELECT "CompanyName", "OrderID" FROM customers \
            JOIN orders ON orders."CustomerID" = customers."CustomerID";
customerdf = result2.DataFrame()

bigdf = customerdf.merge(orderdf, on="OrderID")
bigdf["total"] = bigdf.UnitPrice*bigdf.Quantity - bigdf.Discount
revenue = bigdf.pivot_table(index="CompanyName", values="total", aggfunc=sum)
revue
```
