Scatterplots
------------
**If students score high in one subject, they are likely to score high in the other subject as well.**

```python
fig, (ax1,ax2) = matplotlib.pyplot.subplots(1,2, sharey=True, figsize=(10,4))
ax1.scatter(df.Verbal, df.Math)
seaborn.jointplot(x="Verbal", y="Math", data=df, kind='kde',
                  lowess=True, ratio=10, space=0.1, ax=ax2)
```

**The next two scatter plots show that states that have a low test participation rate, tend to have students that score higher on average.**

```python
fig, (ax1,ax2) = matplotlib.pyplot.subplots(1,2, sharey=True, figsize=(10,4))
seaborn.regplot(x="Rate", y="Math", data=df, lowess=True, ax=ax1)
seaborn.regplot(x="Rate", y="Verbal", data=df, lowess=True, ax=ax2)
```

```python
state_abrv = pandas.read_csv('state_table.csv')
state_abrv
```

```python
state_demos = pandas.read_csv('state_demos.csv')
state_demos
```

```python
df_with_full = df.merge(state_abrv, 
                            left_on='State',
                            right_on='State_Abbreviations',
                            how='left'
                            ).copy()
df_with_full
```

```python
df_complete = df_with_full.merge(state_demos, 
                            left_on='States',
                            right_on='State',
                            how='left'
                            ).copy()
df_complete
```

```python
df_complete['White_Perc'] = df_complete['White Alone'] / df_complete['Total Population']
df_complete['African_Amer_Perc'] = df_complete['Black or African American Alone'] / df_complete['Total Population']
df_complete['Asian_Perc'] = df_complete['Asian Alone'] / df_complete['Total Population']
df_complete['Native_Amer_Perc'] = df_complete['American Indian and Alaska Native Alone'] / df_complete['Total Population']
```

```python
fig, (ax1,ax2,ax3,ax4) = matplotlib.pyplot.subplots(1,4, sharey=True, figsize=(16,5))
seaborn.regplot(x="White_Perc", y="Math", data=df_complete, lowess=True, ax=ax1)
seaborn.regplot(x="African_Amer_Perc", y="Math", data=df_complete, lowess=True, ax=ax2)
seaborn.regplot(x="Asian_Perc", y="Math", data=df_complete, lowess=True, ax=ax3)
seaborn.regplot(x="Native_Amer_Perc", y="Math", data=df_complete, lowess=True, ax=ax4)
```

```python
fig, (ax1,ax2,ax3,ax4) = matplotlib.pyplot.subplots(1,4, sharey=True, figsize=(16,5))
seaborn.regplot(x="White_Perc", y="Verbal", data=df_complete, lowess=True, ax=ax1)
seaborn.regplot(x="African_Amer_Perc", y="Verbal", data=df_complete, lowess=True, ax=ax2)
seaborn.regplot(x="Asian_Perc", y="Verbal", data=df_complete, lowess=True, ax=ax3)
seaborn.regplot(x="Native_Amer_Perc", y="Verbal", data=df_complete, lowess=True, ax=ax4)
```
