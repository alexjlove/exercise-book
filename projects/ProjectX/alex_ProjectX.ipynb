```python
tweet_token = []
for i in range(len(tweets)):
    tweet_token.append(nltk.word_tokenize(tweets[i]))

tweet_token[1]
```

```python
bad_elements = ['#', ':', 'pic.twitter', '@', '&', ';', '<', '>', '.', '/' ,',']
hacked_up = []
hacked_up_list = []
for i in range(len(tweet_token)):
    for j in range(len(tweet_token[i])):
        bad_count = 0
        for e in bad_elements:
            if e in tweet_token[i][j]:
                bad_count = 1
        if bad_count == 0:
            hacked_up.append(tweet_token[i][j])
    hacked_up_list.append(hacked_up)
    hacked_up = []

hacked_up_list
```

```python
positive_file = open("positive-words.txt", "r")
positive_words = positive_file.read().split('\n')
positive_words[35]

negative_file = open('negative-words.txt', 'r')
negative_words = negative_file.read().split('\n')
negative_words[35]

'''
CITATION:
 Minqing Hu and Bing Liu. "Mining and Summarizing Customer Reviews." 
;       Proceedings of the ACM SIGKDD International Conference on Knowledge 
;       Discovery and Data Mining (KDD-2004), Aug 22-25, 2004, Seattle, 
;       Washington, USA
'''
```

```python
positive_words = positive_words[35:]
negative_words = negative_words[35:]
```

```python
def counts(x):
    pos_count = 0
    neg_count = 0
    for word in x:
        if word.lower() in positive_words:
            pos_count +=1
        if word.lower() in negative_words:
            neg_count +=1
        else:
            continue
    score = (pos_count - neg_count)/2.0
    return score

scores = []
for i in range(len(hacked_up_list)):
    scores.append(counts(hacked_up_list[i]))
print scores
total_score = sum(scores)
print total_score
```
